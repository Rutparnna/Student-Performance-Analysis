```python
#Import Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('student-mat.csv',sep=';')
df.head()
```

```python
#Check Dataset Shape

df.shape
```

```python
#check Data Types

df.dtypes
```

```python
#Check Missing values

df.isnull().sum()
```

```python
#Remove Duplicates

df=df.drop_duplicates()
print(df.shape)
```

```python
#Statistical Summary

df.describe()
```

```python
#Average Final Grade(G3)

average_grade=df['G3'].mean()
print("Average Grade:",average_grade)
```

```python
#Students Scoring Above 15

high_score=df[df['G3']>15]
print("students above 15:",len(high_score))
```

```python
#Correlation Between Study Time And Grade

correlation=df['studytime'].corr(df['G3'])
print("Correlation:",correlation)
```

```python
#Gender Performance

gender_avg=df.groupby('sex')['G3'].mean()
print("Gender Performance:")
print(gender_avg)
```

```python
#Hisstogram Of Grades

plt.figure(figsize=(8,4))
plt.hist(df['G3'],bins=10)
plt.title("Distribution Of Final Grades")
plt.xlabel("G3")
plt.ylabel("Number Of Students")
plt.show()
```

```python
#Scatter PLot

plt.figure(figsize=(8,4))
plt.scatter(df['studytime'],df['G3'])
plt.title("Study Time Vs Final Grade")
plt.xlabel("Study Time")
plt.ylabel("Final Grade")
plt.show()
```

```python
#Bar chart

gender_avg.plot(kind='bar')
plt.title("Male vs Female Average Grade")
plt.xlabel("Gender")
plt.ylabel("Average Grade")
plt.show()
```
